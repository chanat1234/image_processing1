```python
frame = cv2.imread(os.path.join(image_path,'ball.png'),cv2.IMREAD_COLOR) #เปิดขึ้นมาเป็นภาพมีสี
img = frame.copy()
plt.axis("off")
#plt.imshow(frame)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Drawing Contours

```python
# Convert BGR to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# define range of color in HSV
lower_color = np.array([29, 86, 6])
upper_color = np.array([64, 255, 255])

# Threshold the HSV image to get only the wanted colors
mask = cv2.inRange(hsv, lower_color, upper_color)

plt.axis("off")
plt.imshow(mask,'gray')
plt.show()
```

```python
contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
n = len(contours)

print('There are %d contours'%n)
```

```python
cv2.drawContours(img, contours, -1, (0,0,255), 20)

plt.axis("off")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
for i in range(0,n):
    cnt = contours[i]
    perimeter = cv2.arcLength(cnt,True)
    print('cnt %2d perimeter is %.0f'%(i,perimeter))
```

## Modify Workshop 3 to draw bounding box on the ball

```python
img = frame.copy()

for i in range(0,n):
    cnt = contours[i]
    perimeter = cv2.arcLength(cnt,True)
    
    if perimeter>500:
      x,y,w,h = cv2.boundingRect(cnt)
      cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),10)

plt.axis("off")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

##Workshop 4: Color Object Tracking

```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
import os
image_path = "/content/gdrive/MyDrive/AI_Course/images"
capture = cv2.VideoCapture(os.path.join(image_path,'ball_tracking.mp4'))

output_file = os.path.join(image_path,'output1.avi')

frame_width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'XVID')
video = cv2.VideoWriter(output_file, fourcc, 25, (frame_width, frame_height))
i = 0

while True:
    has_frame, frame = capture.read()
    i = i+1
    if not has_frame:
        print('Reached the end of the video.')
        print('Processed %d frames.'% i)
        break

    if i==1:
        plt.axis("off")
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.show()
        
    # Convert BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # define range of color in HSV
    lower_color = np.array([29, 86, 6])
    upper_color = np.array([64, 255, 255])
    
    # Threshold the HSV image to get only the wanted colors
    mask = cv2.inRange(hsv, lower_color, upper_color)
    
    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(frame,frame, mask= mask)
    
    video.write(res)
    
capture.release()    
video.release()
```
