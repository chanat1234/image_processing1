```python
# Step 1: Import necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load image
img = cv2.imread("/content/drive/MyDrive/@Downloads/polygon.png")

# Step 3: Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 4: Apply threshold to separate foreground from background
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Step 5: Find contours in the image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Step 6: Loop through the contours and check if they are a circle
for cnt in contours:
    #count angle and set drawContours BGR (Blue-Green-Red)
    approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt, True), True)
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt,True)
    circularity = 4*np.pi*(area/(perimeter*perimeter))
    print(circularity)
    if circularity >= 0.8:
        # Circle
        cv2.drawContours(img, [cnt], 0, (0, 0, 0), 3) 

# Step 7: Display resulting image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Step 1: Import necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load image
img = cv2.imread("/content/drive/MyDrive/@Downloads/polygon_all.png")

# Step 3: Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 4: Apply threshold to separate foreground from background
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Step 5: Find contours in the image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Step 6: Define a dictionary of shapes and their corresponding colors
shapes = {
    3: (255, 0, 0),  # Triangle (blue)
    4: (0, 255, 0),  # Rectangle (green)
    5: (0, 0, 255),  # Pentagon (red)
    6: (255, 255, 0),  # Hexagon (cyan)
    8: (255, 0, 255)  # Octagon (magenta)
}

# Step 7: Loop through the contours and draw the corresponding shape with the appropriate color
for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt, True), True)
    num_sides = len(approx)
    if num_sides in shapes:
        color = shapes[num_sides]
        cv2.drawContours(img, [cnt], 0, color, 3)

# Step 8: Display resulting image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
